### CC 2.1.2: Classes and Object-Oriented Programming

#### Q1: Classes and Object-Oriented Programming

Consider the following code:
    
```
class NewList(list): 
   def remove_max(self): 
     self.remove(max(self)) 
   def append_sum(self): 
     self.append(sum(self)) 

x = NewList([1,2,3]) 
while max(x) < 10: 
   x.remove_max() 
   x.append_sum() 
print(x) 
```

What will this print?

1. 10
2. [5,8,10]
3. Nothing: this code contains an error.
4. **Nothing: this program will never halt.**
