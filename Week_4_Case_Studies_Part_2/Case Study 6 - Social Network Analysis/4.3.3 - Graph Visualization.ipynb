````python
import networkx as nx
import matplotlib.pyplot as plt
````

````python
G = nx.karate_club_graph()
#G.nodes(), G.edges()
G.number_of_nodes(), G.number_of_edges()
````

````python
nx.draw(G, with_labels = 'True', node_color = 'lightblue', edge = 'grey')
plt.savefig('networks_info/karate.pdf')
plt.show(G.nodes(), G.edges())
````

````python
# stores the degres of nodes in a dictionary where the edges are node ids
# and values are their associated degrees
#example node 31 has 6 friends or 6 ties
G.degree()
````
